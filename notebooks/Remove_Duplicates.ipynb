```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
DF_obj = DataFrame({'column 1': [1,1,2,2,3,3,3],
                    'column 2': ['a', 'a', 'b', 'b', 'c', 'c', 'c'],
                    'column 3': ['A', 'A', 'B', 'B','C','C','C']
                    })

DF_obj
```

```python
# .duplicate method
#Dùng để detect dòng dữ liệu (row) bị trùng với dòng dữ liệu trước đó

DF_obj.iloc[1,2] = 'C'
DF_obj.iloc[5,2] = 'X'
DF_obj
```

```python
DF_obj.duplicated()
```

```python
#drop_duplicates

DF_obj.drop_duplicates()
```
